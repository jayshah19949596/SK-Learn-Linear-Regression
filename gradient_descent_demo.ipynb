#  Gradient Descent Demo 

## 0] Import Dependencies 

```python
import numpy as np
import matplotlib.pyplot as plt
import math
% matplotlib inline
```

## 1] Load Data
-------------------------------

```python
def load_data(filename):
    input_file = open(filename, "r")
    first = True
    x_data = []
    y_data = []
    for line in input_file:
        if first:
            x_head, y_head = line.split()
            first = False
            continue
        x_value, y_value = line.split()
        x_data.append(float(x_value))
        y_data.append(float(y_value))
    return x_data, y_data, x_head, y_head
```

## 2] Visualize the Training Data
-------------------------------

```python
def plot_data(x, y, x_label, y_label):
    fig = plt.figure()
    plt.scatter(x, y, figure=fig)
    plt.xlabel(x_label, figure=fig)
    plt.ylabel(y_label, figure=fig)
    plt.show()
```

## 3] Perform Gradient Descent 
-------------------------------

```python
def gradient_descent(x_points, y_points):
    b = 0
    m = 0 
    for i in range(0, iterations):
        b, m = step_gradient(b, m, x_points, y_points)
        plot_line(x_points, y_points, x_heading, y_heading, m, b)
        print("error for iteration", i, "= ", compute_error(b, m, x_points, y_points))
    return [b, m]
```

```python
def plot_line(x, y, x_label, y_label, current_m, current_b):
    fig = plt.figure()
    plt.scatter(x, y, figure=fig)
    plt.xlabel(x_label, figure=fig)
    plt.ylabel(y_label, figure=fig)
    x_values = np.array(x)
    y_values = current_b + (current_m * x_values)
    plt.plot(x_values, y_values, "--", figure=fig)
    plt.show()
```

## 4] Calculate Error Gradient and Update weights 
-----------------------------

 1. N = number of total data points
 2. x = data input 
 3. m = slope of the line
 4. b = y intercept of the line

- f is out error function :
$$ {f}=\frac{1}{N} \sum_{i=1}^{{N}}  [y_{i} - ((m * x_{i}) + b)]^{2}$$



- Error gradient wrt b : 

$$ \frac{df}{db}= - \frac{2}{N} \sum_{i=1}^{{N}}  [y_{i} - ((m * x_{i}) + b)]$$



- Error gradient wrt m : 
$$ \frac{df}{dm}= - \frac{2}{N}  \sum_{i=1}^{{N}} x_{i} * [y_{i} - ((m * x_{i}) + b)]$$


- Weight updation :
                    new_wght = old_wght - (learning_rate * weigt_gradient)

```python
def step_gradient(b_current, m_current, x_values, y_values):
    b_gradient = 0
    m_gradient = 0
    N = float(len(x_values))
    for i in range(0, len(x_values)):
        x = x_values[i]
        y = y_values[i]
        b_gradient += (y - ((m_current * x) + b_current))
        m_gradient += x * (y - ((m_current * x) + b_current))
        
    b_gradient *= - (2 / N) 
    m_gradient *= - (2 / N) 
    new_b = b_current - (learning_rate * b_gradient)
    new_m = m_current - (learning_rate * m_gradient)
    
    return [new_b, new_m]
```

## 5] Compute Error
-------------------------------

- Mean Square Error :
$$ {f}=\frac{1}{N} \sum_{i=1}^{{N}}  [y_{i} - ((m * x_{i}) + b)]^{2}$$


```python
def compute_error(b, m, x_values, y_values):
    total_error = 0
    for i in range(0, len(x_values)):
        x = x_values[i]
        y = y_values[i]
        total_error += (y - (m*x+b))**2
    return total_error/float(len(x_values))
```

```python
file = "data.txt"
iterations = 30
learning_rate = 0.0001
x_points, y_points, x_heading, y_heading = load_data(file)
```

```python
plot_data(x_points, y_points, x_heading, y_heading)
```

```python
optimal_b, optimal_m = gradient_descent(x_points, y_points)
```

```python
print("optimal_b =", optimal_b, "optimal_m =", optimal_m)
```
